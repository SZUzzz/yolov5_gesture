```python
import os

# 设置文件夹路径和新文件名前缀
folder_path = '/home/zz/Pictures/images'  # 替换为你的文件夹路径
new_name_prefix = 'img_1113_'  # 新的文件名前缀
start_index = 1  # 起始编号

# 遍历文件夹中的所有文件
for filename in os.listdir(folder_path):
    # 检查文件是否为图片格式（例如jpg, png, jpeg等）
    if filename.lower().endswith(('.jpg', '.jpeg', '.png', '.gif', '.bmp')):
        # 设置新的文件名
        new_name = f"{new_name_prefix}{start_index}.jpg"  # 修改为你想要的扩展名
        # 生成完整路径dd
        old_path = os.path.join(folder_path, filename)
        new_path = os.path.join(folder_path, new_name)
        
        # 重命名文件
        os.rename(old_path, new_path)
        print(f"Renamed {filename} to {new_name}")
        
        # 更新编号
        start_index += 1

print("Renaming complete!")
```

```python
import os

# 图片和文本文件的文件夹路径
image_folder = '/home/zz/Pictures/images'  # 替换为你的图片文件夹路径
txt_folder = '/home/zz/Pictures/labels'  # 替换为你的文本文件夹路径

# 获取图片和文本文件的文件名（不含扩展名）
image_files = {os.path.splitext(f)[0] for f in os.listdir(image_folder) if f.lower().endswith(('.jpg', '.jpeg', '.png', '.gif', '.bmp'))}
txt_files = {os.path.splitext(f)[0] for f in os.listdir(txt_folder) if f.lower().endswith('.txt')}

# 找出没有对应 txt 文件的图片文件
images_without_txt = image_files - txt_files

# 删除没有对应 txt 文件的图片
for image_name in images_without_txt:
    # 构建图片的完整路径（假设图片是 .jpg，若有其他扩展名需要自行修改）
    image_path = os.path.join(image_folder, image_name + '.jpg')
    
    if os.path.exists(image_path):
        os.remove(image_path)
        print(f"Deleted {image_path}")
    else:
        print(f"{image_path} not found. Skipping.")

print("Operation complete!")
```

```python
import os

# 设置文本文件夹路径
txt_folder = '/home/zz/Pictures/labels'  # 替换为你的文本文件夹路径

# 定义替换规则
replace_dict = {
    '15': '4',
    '16': '2',
    '17': '1',
    '18': '3',
    '19': '0',
    '20': '5'
}

# 遍历所有 txt 文件
for txt_file in os.listdir(txt_folder):
    if txt_file.endswith('.txt'):
        txt_path = os.path.join(txt_folder, txt_file)

        # 读取文件内容
        with open(txt_path, 'r') as file:
            lines = file.readlines()

        # 修改第一个数字
        modified_lines = []
        for line in lines:
            # 分割行内容，找到第一个数字并替换
            parts = line.split()
            if parts and parts[0] in replace_dict:
                parts[0] = replace_dict[parts[0]]
            # 重新组合行内容
            modified_lines.append(' '.join(parts) + '\n')

        # 将修改后的内容写回文件
        with open(txt_path, 'w') as file:
            file.writelines(modified_lines)

        print(f"Modified {txt_file}")

print("Operation complete!")
```

```python
import os

# 图片和文本文件的文件夹路径
image_folder = '/home/zz/yolov5/shoushi_data/images/val'  # 替换为你的图片文件夹路径
txt_folder = '/home/zz/yolov5/shoushi_data/labels/val'  # 替换为你的文本文件夹路径

# 获取并排序图片和文本文件的文件名（不含扩展名）
image_files = sorted([f for f in os.listdir(image_folder) if f.lower().endswith(('.jpg', '.jpeg', '.png', '.gif', '.bmp'))])
txt_files = sorted([f for f in os.listdir(txt_folder) if f.lower().endswith('.txt')])

# 去掉扩展名进行匹配
image_files_no_ext = [os.path.splitext(f)[0] for f in image_files]
txt_files_no_ext = [os.path.splitext(f)[0] for f in txt_files]

# 确保两者顺序一致（并打印结果）
print("Sorted Image Files:", image_files_no_ext)
print("Sorted Txt Files:", txt_files_no_ext)

# 检查每个文件是否一一对应
for img, txt in zip(image_files_no_ext, txt_files_no_ext):
    if img == txt:
        print(f"Match: {img} -> {txt}")
    else:
        print(f"Mismatch: {img} does not match {txt}")

print("Sorting and matching complete!")
```
